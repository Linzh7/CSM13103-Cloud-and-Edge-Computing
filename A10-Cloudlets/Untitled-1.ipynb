```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
df = pd.read_csv('ping.csv', sep=',')
```

```python
df.head()
```

```python
array = df.to_numpy()
```

```python
label_list = ['min', 'max', 'avg']
time_list = ['17.2.2023 DNA NIGHT', '18.2.2023 DNA MORNING', '18.2.2023 Elisa MORNING', '19.2.2023 DNA NOON']
line_list = [
    'www.google.com', 'www.google.fi', 'www.yle.fi', 'www.hbomax.com',
    'www.ntnu.edu', 'www.inria.fr', 'www.washington.edu', 'www.inria.cl'
]
plot_data = defaultdict(list)
for label in label_list:
    for i in array:
        # print(i)
        if i[1] == label:
            plot_data[label].append(list(i[2:]))
    #     break
    # break
print(plot_data)
```

```python
tmp_list = np.array(plot_data['avg']).transpose().tolist()
```

```python
for i in range(8):
    x = np.array([0,1,2,3])
    plt.xticks(x, time_list,rotation=10)
    plt.plot(tmp_list[i])
    plt.legend(line_list, bbox_to_anchor=(1, 1))
plt.savefig('avg.png', bbox_inches='tight')
```
